```python
import numpy as np
from PIL import Image
import os

# directory containing the image files
normal_path = '/content/gdrive/MyDrive/GeneratedPhotosNormal'

# set the desired size for the resized images
size = (128, 128)

# create an empty list to store the image data
image_data = []

# loop through the image file paths and load each image
for file_name in os.listdir(normal_path):
    # check if the file is an image file
    if file_name.endswith('.jpg') or file_name.endswith('.png'):
        # create the full file path
        file_path = os.path.join(normal_path, file_name)
        # open the image file using PIL
        img = Image.open(file_path)
        # convert the image to RGB mode if it is not already in RGB mode
        if img.mode != 'RGB':
            img = img.convert('RGB')
        # resize the image to the desired size
        img = img.resize(size)
        # convert the PIL image to a NumPy array
        img_array = np.array(img)
        # append the image data to the list
        image_data.append(img_array)

# convert the list of image data to a NumPy array
image_array_normal = np.array(image_data)

# print the shape of the image array
print(image_array_normal.shape)
```

```python
import numpy as np
from PIL import Image
import os

# directory containing the image files
normal_path = '/content/gdrive/MyDrive/GeneratedPhotosABNormal'

# set the desired size for the resized images
size = (128, 128)

# create an empty list to store the image data
image_data = []

# loop through the image file paths and load each image
for file_name in os.listdir(normal_path):
    # check if the file is an image file
    if file_name.endswith('.jpg') or file_name.endswith('.png'):
        # create the full file path
        file_path = os.path.join(normal_path, file_name)
        # open the image file using PIL
        img = Image.open(file_path)
        # convert the image to RGB mode if it is not already in RGB mode
        if img.mode != 'RGB':
            img = img.convert('RGB')
        # resize the image to the desired size
        img = img.resize(size)
        # convert the PIL image to a NumPy array
        img_array = np.array(img)
        # append the image data to the list
        image_data.append(img_array)

# convert the list of image data to a NumPy array
image_array_ABnormal = np.array(image_data)

# print the shape of the image array
print(image_array_ABnormal.shape)
```

```python
import os
import matplotlib.pyplot as plt
plt.imshow(image_array_ABnormal[6])
plt.axis('off')
plt.show()
```

```python
imageNormalizedAbNormal = image_array_ABnormal / 255.0
print(imageNormalizedAbNormal)
```

```python
import numpy as np
from PIL import Image

#imageNormalizedNormal
#imageNormalizedAbNormal

# Create labels for the images
normal_baby_labels = np.zeros(imageNormalizedNormal.shape[0])  # 0 for normal babies
abnormal_baby_labels = np.ones(imageNormalizedAbNormal.shape[0])  # 1 for abnormal babies

# Concatenate the images and labels into a single dataset
data = np.concatenate((imageNormalizedNormal, imageNormalizedAbNormal))
labels = np.concatenate((normal_baby_labels, abnormal_baby_labels))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, labels, test_size = 0.25,random_state=42)
```

```python
y_train.shape
```

```python
from tensorflow import keras
from tensorflow.keras.models import load_model

# Load the saved model from files
model = load_model('/content/gdrive/MyDrive/model/trained_model_new.h5')

# Evaluate the model on the test data
print(model.evaluate(X_test, y_test))

```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPool2D
from keras.optimizers import RMSprop
from keras.preprocessing.image import ImageDataGenerator
from keras.callbacks import ReduceLROnPlateau

model = Sequential()

model.add(Conv2D(filters = 32, kernel_size = (5,5),padding = 'Same',
                 activation ='relu', input_shape = (128,128,3)))
model.add(Conv2D(filters = 32, kernel_size = (5,5),padding = 'Same',
                 activation ='relu'))
model.add(MaxPool2D(pool_size=(2,2)))
model.add(Dropout(0.25))


model.add(Conv2D(filters = 64, kernel_size = (3,3),padding = 'Same',
                 activation ='relu'))
model.add(Conv2D(filters = 64, kernel_size = (3,3),padding = 'Same',
                 activation ='relu'))
model.add(MaxPool2D(pool_size=(2,2), strides=(2,2)))
model.add(Dropout(0.25))

model.add(Conv2D(filters = 128, kernel_size = (1,1),padding = 'Same',
                 activation ='relu'))
model.add(Conv2D(filters = 128, kernel_size = (1,1),padding = 'Same',
                 activation ='relu'))
model.add(MaxPool2D(pool_size=(2,2), strides=(2,2)))
model.add(Dropout(0.25))

model.add(Flatten())
model.add(Dense(600, activation = "relu"))
model.add(Dropout(0.5))
model.add(Dense(1, activation='sigmoid'))
model.compile(optimizer = "Adam" , loss = "binary_crossentropy", metrics=["accuracy"])
```

```python
model.fit(X_train, y_train, epochs=25)
```

```python


model.save("/content/gdrive/MyDrive/model/trained_model_new.h5")

# Evaluate the model on the test set
#loss, accuracy = model.evaluate(X_test, y_test)
#print(f'Test loss: {loss}, test accuracy: {accuracy}')


```

```python
from sklearn.metrics import classification_report

# Make predictions on the test data
y_pred = model.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=1)

# Generate a classification report
cr = classification_report(y_test, y_pred_classes)

# Print the classification report
print(cr)
```
